```python
## Time Resampling
```

```python
import pandas as pd
```

```python
df = pd.read_csv('Data/starbucks.csv',index_col = 'Date',parse_dates = True)
```

```python
df
```

```python
df.info()
```

```python
#Without parse_dates = True the dtype of index i.e Date will be of object
##After parse_dates = True the dtype of index is now datetime64[ns]
df.index
```

```python
# Resample daily data into yearly data
df.resample(rule='A')
```

```python
df.resample(rule='A').mean()
```

```python
df.resample(rule='A').std()
```

```python
def first_day(entry) :

    #is their an entry?
    if len(entry) != 0:
        #if so,return first entry
        return entry[0]
```

```python
df.resample(rule='A').apply(first_day)
```

```python
df.head()
```

```python
df
```

```python
df['Close'].resample('A').mean().plot.bar()
```

```python
#Time Shifting
"""We get an empty data row in the head and in the tail we lose one data row ie. we lose valuable information"""
```

```python
df.shift(1).tail ()
```

```python
df.shift(5)
```

```python
df.shift(1)
```

```python
"""We end up losing one point at above"""

df.shift(-1)
```

```python
df.head()
```

```python
df.shift(periods=1,freq='M')
```
